### Importing Necessary Libraries

```python
import sqlite3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

### Establishing the Connection with Sakila DB File

```python
#Establish a connection to the SQLite database
connection = sqlite3.connect('sakila.db')
```

### Question-1: Movies Count as per Duration / Length

```python
#Define SQL query
sql_query = """
	SELECT 
	CASE 
		WHEN length < 50 THEN 'Short_Film'
		WHEN length Between 50 AND 100 THEN 'Medium_Length'
		ELSE 'Long_Film'
		END AS 'Film_Duration',
		COUNT(*) AS 'Film_Count'
		FROM film
	GROUP BY Film_Duration;
"""
```

```python
# Using Pandas to execute the SQL Query and save results in Data-Frame
df_films_count_duration = pd.read_sql_query(sql_query, connection)
```

```python
df_films_count_duration.head()
```

```python
# Create the pie plot
plt.pie(df_films_count_duration['Film_Count'], labels=df_films_count_duration.Film_Duration, autopct='%1.1f%%', startangle=90)

# Add a circle at the center to create the donut effect
my_circle = plt.Circle((0, 0), 0.7, color='white') # (0,0) is center, 0.7 is radius
p = plt.gcf() # Get the current figure
p.gca().add_artist(my_circle) # Get current axes and add the circle artist

plt.title('Film Count Per Duration Category')
plt.axis('equal')
plt.show()
```

### Question-2: Monthly Payment Received by Each Staff

```python
#Define SQL query
sql_query = """
	SELECT
		TB1.Month, concat(TB1.first_name," ",TB1.last_name) AS Staff_Name ,cast(sum(TB1.amount) AS INT) AS Payment_BY_Staff
	FROM
		(
		SELECT 
			I.inventory_id,I.film_id,r.rental_id,p.staff_id,p.amount,STRFTIME('%Y-%m', p.payment_date) AS Month ,S.first_name,S.last_name
		FROM inventory I
		LEFT JOIN rental r 
		ON I.inventory_id = r.inventory_id
		JOIN payment p
		ON r.rental_id = p.rental_id
		JOIN staff S
		ON p.staff_id=S.staff_id
		)TB1
	GROUP BY TB1.Month, TB1.staff_id
	ORDER BY TB1.Month,TB1.staff_id;
"""
```

```python
# Using Pandas to execute the SQL Query and save results in Data-Frame
df_monthly_payment_staff = pd.read_sql_query(sql_query, connection)
```

```python
df_monthly_payment_staff
```

```python
# Create a line plot with 'event' as the hue category
sns.lineplot(x="Month", y="Payment_BY_Staff", hue="Staff_Name", data=df_monthly_payment_staff,markers=True,marker='o', markersize=10)
# Add title and show plot
plt.title("Monthy Received Payment by Staff")
plt.show()
```

### Question-3: Top 10 Actors by Films Count

```python
#Define SQL query
sql_query = """
	WITH film_by_Actors AS
		(
		SELECT 
			f.film_id,fa.actor_id,concat(a.first_name," ", a.last_name) AS Actor_Name
			FROM film f
			JOIN film_actor fa
			ON f.film_id = fa.film_id
			JOIN actor a
			ON fa.actor_id = a.actor_id
		)
		
	SELECT
		Actor_Name,
		COUNT(*) AS Films_Count
	FROM
		film_by_Actors
	GROUP BY Actor_Name
	ORDER BY COUNT(*) DESC
	LIMIT 10;
"""
```

```python
# Using Pandas to execute the SQL Query and save results in Data-Frame
df_Top10_Actors = pd.read_sql_query(sql_query, connection)
```

```python
df_Top10_Actors
```

```python
df_Top10_Actors.sort_values(by='Films_Count').plot.barh(x='Actor_Name', y='Films_Count',
             title='Top-10 Actors by Film Count', color='green')
plt.show()
```

### Question-4: Total Rental Amount per Film category

```python
#Define SQL query
sql_query = """
	WITH CTE AS 
		(
		SELECT r.rental_id,r.inventory_id,p.amount,i.film_id,f.title,fc.category_id,c.name
		FROM rental r
		LEFT JOIN payment p
		ON r.rental_id = p.rental_id
		LEFT JOIN inventory i
		ON r.inventory_id = i.inventory_id
		LEFT JOIN film f
		ON i.film_id = f.film_id
		LEFT JOIN film_category fc
		ON f.film_id = fc.film_id
		LEFT JOIN category c
		ON fc.category_id = c.category_id
	)

	SELECT 
		CTE.name AS Film_Category, 
		SUM(CTE.amount) AS Total_rental_amount
	FROM CTE
	GROUP BY CTE.name
	ORDER BY SUM(CTE.amount) DESC;
"""
```

```python
# Using Pandas to execute the SQL Query and save results in Data-Frame
df_rental_amount_category = pd.read_sql_query(sql_query, connection)
```

```python
df_rental_amount_category
```

```python
plt.figure(figsize=(14, 4))
sns.barplot(df_rental_amount_category, x="Film_Category", y="Total_rental_amount")
plt.title('Rental Amount by Film Categories')
plt.show()
```

### Question-5: Films Percentage Count per Rating Category

```python
#Define SQL query
sql_query = """
	SELECT
		rating FilmRating,
		COUNT(*) AS CategoryCount,
		(COUNT(*) * 100.0 / SUM(COUNT(*)) OVER()) || '%'  AS PercentageOfTotal
	FROM
		film
	GROUP BY
		rating
	ORDER BY
		PercentageOfTotal DESC;
"""
```

```python
# Using Pandas to execute the SQL Query and save results in Data-Frame
df_Film_percentage_rating = pd.read_sql_query(sql_query, connection)
```

```python
df_Film_percentage_rating
```

```python
myexplode = [0.2, 0, 0, 0, 0]
plt.pie(df_Film_percentage_rating.CategoryCount, labels = df_Film_percentage_rating.FilmRating, explode = myexplode, shadow = True,autopct='%0.0f%%')
plt.title('Percentage of Film Counts per Rating Category')
plt.show() 
```
